## Hysteresis and power

This notebook generates Figures 5 and 6 from the paper. Just execute the code cell. 

```python
import random
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.cm as cm
import numpy as np
import seaborn as sns
%matplotlib inline

def calculate_probability(t, max_time=400, power=2):
    probability = (t / max_time) ** power
    return min(probability, 1.0)  # Ensure the probability does not exceed 1

def simulate_decision(max_time=400, power=2):
    for t in range(1, max_time + 1):
        probability = calculate_probability(t, max_time, power)
        if random.random() < probability:
            return t
    return max_time  # In case no decision is made, return max_time

def run_simulation(trials=1000, max_time=400, power=2):
    decision_times = []
    for _ in range(trials):
        decision_time = simulate_decision(max_time, power)
        decision_times.append(decision_time)
    return decision_times

def get_color_for_power(power):
    if power == 1:
        return 'purple'  # Use purple for power = 1
    elif power < 1:
        # Normalize power values in the range [0.125, 0.5] to [0, 1]
        normalized_power = (power - 0.125) / (0.5 - 0.125)  # Normalize to [0, 1]
        return mcolors.to_hex(mcolors.LinearSegmentedColormap.from_list("", ["blue", "purple"])(normalized_power))
    else:
        # Normalize power values in the range [2, 8] to [0, 1]
        normalized_power = (power - 2) / (8 - 2)  # Normalize to [0, 1]
        return mcolors.to_hex(mcolors.LinearSegmentedColormap.from_list("", ["purple", "red"])(normalized_power))

def plot_decision_times(decision_times, ax, mt, power):
    bins = np.arange(1, max(decision_times) + 2)  # Bins of exactly 1
    counts, bin_edges = np.histogram(decision_times, bins=bins)
    bin_centers = bin_edges[:-1] + 0.5  # Center of each bin
    
    # Filter out points where frequency is 0
    nonzero_indices = counts > 0
    bin_centers = bin_centers[nonzero_indices]
    counts = counts[nonzero_indices]

    ax.plot(bin_centers, counts, 'o', markersize=1)  # Plot with points
    # sns.kdeplot(decision_times, ax=ax, color='red')  # Superimpose density plot

    ax.set_title(f'Decision Times Frequency\nmt={mt}, power={power}')
    ax.set_xlabel('Timestep')
    ax.set_ylabel('Frequency')
    
    # Annotate with min, max, and mean
    min_time = np.min(decision_times)
    max_time = np.max(decision_times)
    mean_time = np.mean(decision_times)
    ax.annotate(f'Min: {min_time}', xy=(0.75, 0.95), xycoords='axes fraction', ha='left', va='top', fontsize=10, color='blue')
    ax.annotate(f'Max: {max_time}', xy=(0.75, 0.90), xycoords='axes fraction', ha='left', va='top', fontsize=10, color='blue')
    ax.annotate(f'Mean: {mean_time:.2f}', xy=(0.75, 0.85), xycoords='axes fraction', ha='left', va='top', fontsize=10, color='blue')
    
    # Set y-axis to show only integer values
    ax.yaxis.set_major_locator(plt.MaxNLocator(integer=True))
    ax.set_xlim(-5, 805)  # Set x-axis limits to 0-1200

def find_constant(hysteresis, power):
    x = np.arange(1, hysteresis + 1)
    y = (x / hysteresis) ** power
    C = 1 / np.sum(y)
    print(f'Constant is {C}.')
    return C
    
def plot_decision_function_multiple_powers(max_time=400, power_values=None, ax=None, show_legend=False, rescale=False):
    linestyle_styles = [
         ('densely dashed',        (0, (5, 1))),    
         ('dotted',                (0, (1, 1))),
         ('dashed',                (0, (5, 5))),
         ('densely dotted',        (0, (1, 1))),
         ('solid', 'solid'),      # Same as (0, ()) or '-`'
         ('densely dotted',        (0, (1, 1))),
         ('dashed',                (0, (5, 5))),
         ('dotted',                (0, (1, 1))),
         ('densely dashed',        (0, (5, 1))),    
    ]
    
    if power_values is None:
        power_values = [0.125, 0.167, 0.25, 0.5, 1, 2, 4, 6, 8]
    
    timesteps = range(1, max_time + 1)
    
    for idx, power in enumerate(power_values):
        probabilities = [calculate_probability(t, max_time, power) for t in timesteps]
        
        if rescale:
            constant = find_constant(max_time, power)  # Get the constant based on current max_time and power
            probabilities = [p * constant for p in probabilities]  # Rescale the probabilities
        
        style = linestyle_styles[idx % len(linestyle_styles)][1]  # Use the line style tuple
        color = get_color_for_power(power)  # Get the color based on power
        ax.plot(timesteps, probabilities, linestyle=style, color=color, label=f's={power}')  # Apply the linestyle and color
    
    ax.set_title(f'h={max_time}', fontsize=20)
    ax.set_xlabel('Timestep', fontsize=20)
    ax.set_ylabel('Probability', fontsize=20)
    ax.grid(True)
    
    # Set tick label font size to 18
    ax.tick_params(axis='both', which='major', labelsize=18)

    if show_legend:
        ax.legend(fontsize=18)
    
    ax.set_xlim(-30, 810)  # Set x-axis limits to 0-1200

def savplot_decision_function_multiple_powers(max_time=400, power_values=None, ax=None, show_legend=False, rescale=False):
    linestyle_styles = [
         ('densely dashed',        (0, (5, 1))),    
         ('dotted',                (0, (1, 1))),
         ('dashed',                (0, (5, 5))),
         ('densely dotted',        (0, (1, 1))),
         ('solid', 'solid'),      # Same as (0, ()) or '-`'
         ('densely dotted',        (0, (1, 1))),
         ('dashed',                (0, (5, 5))),
         ('dotted',                (0, (1, 1))),
         ('densely dashed',        (0, (5, 1))),    
    ]
    
    if power_values is None:
        power_values = [0.125, 0.167, 0.25, 0.5, 1, 2, 4, 6, 8]
    
    timesteps = range(1, max_time + 1)
    
    for idx, power in enumerate(power_values):
        probabilities = [calculate_probability(t, max_time, power) for t in timesteps]
        style = linestyle_styles[idx % len(linestyle_styles)][1]  # Use the line style tuple
        color = get_color_for_power(power)  # Get the color based on power
        ax.plot(timesteps, probabilities, linestyle=style, color=color, label=f's={power}')  # Apply the linestyle and color
    
    ax.set_title(f'h={max_time}', fontsize=20)
    ax.set_xlabel('Timestep', fontsize=20)
    ax.set_ylabel('Probability', fontsize=20)
    ax.grid(True)
    # Set tick label font size to 18
    ax.tick_params(axis='both', which='major', labelsize=18)

    if show_legend:
        ax.legend(fontsize=18)
    
    ax.set_xlim(-30, 810)  # Set x-axis limits to 0-1200


# Set up the figure with 4 columns and 6 rows (one row for each mt value)
fig, axs = plt.subplots(2, 2, figsize=(12, 10))  # Set up 2x2 plot
mt_values = [200, 400, 600, 800]  # Four mt values
power_values = [4]  # Only one power value for simplicity in the frequency plot
probability_powers = [0.125, 0.167, 0.25, 0.5, 1, 2, 4, 6, 8]
nsim = 1000

for i, mt in enumerate(mt_values):
    # Use axs[i//2, i%2] to map the 4 mt values to the 2x2 grid
    show_legend = (i == 0)  # Show legend only for the first subplot
    plot_decision_function_multiple_powers(max_time=mt, power_values=probability_powers, ax=axs[i//2, i%2], show_legend=show_legend)
    
    # # Loop through the columns for the decision time frequency plots with different powers
    # for j, power in enumerate(power_values):
    #     decision_times = run_simulation(trials=nsim, max_time=mt, power=power)
    #     plot_decision_times(decision_times, ax=axs[i, j+1], mt=mt, power=power)

plt.tight_layout()
plt.savefig('PNG/fig05_hysteresis_and_power.png')
plt.show()




# Set up the figure with 4 columns and 6 rows (one row for each mt value)
fig, axs = plt.subplots(2, 2, figsize=(12, 10))  # Set up 2x2 plot
mt_values = [200, 400, 600, 800]  # Four mt values
power_values = [4]  # Only one power value for simplicity in the frequency plot
probability_powers = [0.125, 0.167, 0.25, 0.5, 1, 2, 4, 6, 8]
nsim = 1000

for i, mt in enumerate(mt_values):
    # Use axs[i//2, i%2] to map the 4 mt values to the 2x2 grid
    show_legend = (i == 0)  # Show legend only for the first subplot
    plot_decision_function_multiple_powers(max_time=mt, power_values=probability_powers, ax=axs[i//2, i%2], show_legend=show_legend, rescale=True)
    
    # # Loop through the columns for the decision time frequency plots with different powers
    # for j, power in enumerate(power_values):
    #     decision_times = run_simulation(trials=nsim, max_time=mt, power=power)
    #     plot_decision_times(decision_times, ax=axs[i, j+1], mt=mt, power=power)

plt.tight_layout()
plt.savefig('PNG/fig06_hysteresis_and_power_rescaled.png')
plt.show()
```
